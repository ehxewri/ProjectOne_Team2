```python
# read the file into a list
list_of_series_df = pd.read_csv('C:\SRC\AI\ProjectOne_Team2\\resources\\filename.csv')
display (list_of_series_df)
```

```python
series_list_df = {}
# read the file into a list
for index, row in list_of_series_df.iterrows():
    file_id=list_of_series_df.loc[index,'filename']
    county=list_of_series_df.loc[index,'county']
    desc=list_of_series_df.loc[index,'series_Desc']
    series_id = file_id + '_pd'
    series_list_df[series_id] =  pd.read_csv(f'C:\SRC\AI\ProjectOne_Team2\\resources\{file_id}.csv',
                                index_col='date',
                                parse_dates=True,
                                date_format='%Y-%m-%d')
    series_list_df[series_id].plot()  # This plots all columns with numerical data against the index
    plt.title(f'Data from {file_id}')  # Set the title of the plot to include the filename
    plt.xlabel('Year')  # Set the x-axis label
    plt.ylabel(desc)  # Set the y-axis label
    plt.savefig(f'../reports/{county} {desc}.png')
```
